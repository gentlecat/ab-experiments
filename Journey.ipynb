```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import os
import simplejson

all_files = []
for root, dirs, files in os.walk('data/journey'):
    for file in files:
        all_files.append(os.path.join(root, file))

def parse_track_data(path):
    if path.endswith('.json'):
        with open(path, 'r') as f:
            json = simplejson.load(f)            
            return os.path.split(path)[-1][:-5], json

tracks = map(parse_track_data, all_files)
track_count = len(tracks)

tracks.sort(key=lambda pair: int(pair[0]))  # sorting by name (should be a track number)
```

```python
data = [t[1]['lowlevel']['spectral_energy']['mean'] for t in tracks]
plt.plot(data, 'mx')
plt.title('spectral_energy (mean)')
plt.xlabel('track #')
plt.axis([0, track_count, 0, max(data) * 1.5])
plt.show()
```

```python
data = [t[1]['rhythm']['bpm'] for t in tracks]
plt.plot(data, 'rx')
plt.title('beats_count')
plt.xlabel('track #')
plt.axis([0, track_count, 0, max(data) * 1.5])
plt.show()
```

```python
data = [t[1]['rhythm']['danceability'] for t in tracks]
plt.plot(data, 'x')
plt.title('danceability')
plt.xlabel('track #')
plt.axis([0, track_count, 0, 1])
plt.show()
```
